```python
amortized_time_per_attack = {}

for key, value in time_for_n_attacks.items():
    amortized_time_per_attack[key] = {k: v  for k, v in value.items()}

# Plot the amortized time per attack
plt.figure(figsize=(12, 8))
for method, values in amortized_time_per_attack.items():
    x = list(values.keys())
    y = list(values.values())
    plt.plot(x, y, marker='o', label=method)

plt.xlabel("Number of Attacks")
plt.ylabel("Amortized Time per Attack")
plt.title("Amortized Time per Attack (Original Data)")
plt.legend()
plt.grid()
plt.show()
```
